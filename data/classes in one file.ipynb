```python
import numpy as np
import os
import matplotlib.pyplot as plt
from utils import create_img_from_classes
import json
```

```python
dirs = ["test/","train/","val/"]
```

```python
for root_dir in dirs:
    full_dir = root_dir + "lc_classes/"
    files = os.listdir(full_dir)
    i = 0
    d = {}
    for file in files:
        name = file.split(".")[0]
        fn = full_dir + file
        with np.load(fn) as classes:
            d[name] = classes["arr_0"]

        if i % 200 == 0:
            print(f"{i} / {len(files)}", end=" ")
        i += 1
    np.savez_compressed(root_dir + "classes", **d)
```
